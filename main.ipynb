```python
from pathlib import Path
import tifffile as tiff
import numpy as np

folder = Path("/media/ramyar/New Volume/PACS445 220kv95ua8fps24dbdg1e20i-400fr4O [2024-05-30 14.21.51]") 
tif_files = sorted(folder.glob("*.tif"))

print("Number of slices:", len(tif_files))

# Read first slice to get shape and dtype
sample = tiff.imread(tif_files[0])
H, W = sample.shape
dtype = sample.dtype

volume = np.empty((len(tif_files), H, W), dtype=dtype)

for i, f in enumerate(tif_files):
    volume[i] = tiff.imread(f)

print("Volume shape (Z, Y, X):", volume.shape)
print("Data type:", volume.dtype)
```
