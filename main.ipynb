```python

dtype = np.dtype("<u2")  # little-endian uint16

vol = np.memmap(raw_path, dtype=dtype, mode="r", shape=(NZ, NY, NX))  # (Z,Y,X)

print(vol.shape, vol.dtype)  # (2794, 1522, 1650) uint16
```

```python
def save_png(img16, path):
    img = img16.astype(np.float32)
    img = (img - img.min()) / (img.max() - img.min() + 1e-8)
    img8 = (img * 255).astype(np.uint8)
    imageio.imwrite(path, img8)
```

```python
out = Path("front_view_png")
out.mkdir(exist_ok=True)

for z in range(NZ):
    front = vol[z, :, :]   # (Y, X)
    save_png(front, out / f"front_{z:04d}.png")

    if z % 50 == 0:
        print(f"Saved {z}/{NZ}")
```
